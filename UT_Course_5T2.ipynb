```python
### Task 2 will require loading in data and cleaning the data

#import section
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read in csv file
credit = pd.read_csv('default of credit card clients.csv', header = 1)
credit.head()
```

```python
credit.describe()
```

```python
credit.info()
```

```python
### Clean up the data
# SEX, EDUCATION AND MARRIAGE as well as Pay 0:Pay 6 need to be categorical via pandas
credit['SEX'] = credit['SEX'].astype('category')
credit['EDUCATION'] = credit['EDUCATION'].astype('category')
credit['MARRIAGE'] = credit['MARRIAGE'].astype('category')
credit['PAY_0'] = credit['PAY_0'].astype('category')
credit['PAY_2'] = credit['PAY_2'].astype('category')
credit['PAY_3'] = credit['PAY_3'].astype('category')
credit['PAY_4'] = credit['PAY_4'].astype('category')
credit['PAY_5'] = credit['PAY_5'].astype('category')
credit['PAY_6'] = credit['PAY_6'].astype('category')
credit.dtypes
```

```python
print(credit.iloc[:, 12:24])
```

```python
# Delete observations based on no activity - sum up bill and pay amounts. If they = 0, then deleted row
creditBillPay = credit.iloc[:, 12:24]
creditSum = creditBillPay.sum(axis = 1, skipna = True)
print(credit.iloc[:, 12:24])
```

```python
# get index of all rows that sum to zero indicating no activity on the account so I can delete it later
list_zero = creditSum.index[creditSum == 0]
creditDrop = credit
creditDrop.drop(list_zero, inplace=True)

#verify that rows of not activity have been deleted
print(creditDrop.iloc[:, 12:24])
```

```python
#Drop the ID Column
creditDrop = creditDrop.drop(['ID'], axis = 1)
creditDrop.head()
```
