```python
########################################
#### UT Course 5 Task 3 - Regression ###
########################################

#imports: numpy, pandas, scipy, math, matplotlib
import numpy as np
import pandas as pd
import scipy
from math import sqrt
import matplotlib.pyplot as plt

#import estimators:
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn import linear_model

#import: model matrices
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

#import: cross validation
from sklearn.model_selection import train_test_split
```

```python
#import cleaned data from Task 5.2
rawData = pd.read_csv('cleaned_credit_data.csv')
```

```python
### Clean up the data
#rawData['age_by_decade'] = rawData['age_by_decade'].astype('category')
rawData = rawData.drop(['Unnamed: 0'], axis = 1)
```

```python
#rawData information
rawData.info()
#rawData.columns
```

```python
#find column index of dependent variable
#print(rawData.iloc[:, 16])
```

```python
#################### Machine Learning Time ########################
# Establish variables for three different models that I'll be using
modelSVR = SVR()
modelRF = RandomForestRegressor()
modelLR = LinearRegression()

# dependent variable
depVar = rawData.iloc[:, 16]
X = rawData
X = rawData.drop(['default'], axis = 1)
```

```python
# Split the data into train/test - 70/30
X_train, X_test, y_train, y_test = train_test_split(X, depVar, test_size=0.3, random_state=333)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# Assign data to model variables
#modelSVR.fit(X_train, y_train)
#modelRF.fit(X_train, y_train)
#modelRF.fit(X_train, y_train)
```

```python
# Test out a single model to make sure X,y split works
#print(cross_val_score(modelRF, X_train, y_train, cv=5, n_jobs=3))
```

```python
# RF model fitting and scoring
modelRF.fit(X_train,y_train)
print(cross_val_score(modelRF, X_train, y_train, cv=5, n_jobs=3))
modelRF.score(X_train,y_train)

#SVM model fitting and scoring
modelSVR.fit(X_train,y_train)
print(cross_val_score(modelSVR, X_train, y_train, cv=5, n_jobs=3)) 
modelSVR.score(X_train,y_train)

#LM model fitting and scoring
modelLR.fit(X_train,y_train)
print(cross_val_score(modelLR, X_train, y_train, cv=5, n_jobs=3)) 
modelLR.score(X_train,y_train)
```

```python
# Predict the results against the test data
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

predictions = modelLR.predict(X_test)
rmse = sqrt(mean_squared_error(y_test, predictions))
pred_R_squared = r2_score(y_test, predictions)
```

```python
#Print out the results: 
print('R Squared: %.3f' % pred_R_squared)
print('RMSE: %.3f' % rmse)
```
